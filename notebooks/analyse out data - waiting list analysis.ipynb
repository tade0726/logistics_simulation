```python
import sys
sys.path.append("..")

import os
from os.path import join, splitext

import pandas as pd
from src.db import *
from src.config import *
```

```python
def reading_tables():
    table_names = os.listdir(SaveConfig.OUT_DIR)
    tables_dict = dict()
    for name in table_names:
        tables_dict[splitext(name)[0]] = pd.read_csv(join(SaveConfig.OUT_DIR, name))
    return tables_dict
```

```python
machine_table.real_time_stamp = pd.to_datetime(machine_table.real_time_stamp)
```

```python
machine_table_wait = machine_table[machine_table.action == 'wait']
machine_table_start = machine_table[machine_table.action == 'start']
```

```python
machine_table_wait.head()
```

```python
def ret_time_series(table: pd.DataFrame):
    table_temp = table.groupby(['equipment_id', 'real_time_stamp'])['package_id'].value_counts()
    return table_temp.groupby('equipment_id').apply(lambda x: x.resample('s', level=1).sum()).fillna(0)
```

```python
table_wait = ret_time_series(machine_table_wait)
table_start = ret_time_series(machine_table_start)
```

```python
table_total = table_wait.sub(table_start, fill_value=0)
table_total = table_total.fillna(0)
table_final = table_total.groupby(level=0).cumsum()
```

```python
equipement_groups = table_final.groupby(level=0)
```

```python
table_r5_4 = equipement_groups.get_group('r5_4')
table_r5_1 = equipement_groups.get_group('r5_1')
table_r1_1 = equipement_groups.get_group('r1_1')
```
