```python
import pandas as pd
import numpy as np
import pickle
```

```python
def load_object(file_path):
    try:
        with open(file_path, "rb") as file:
            obj = pickle.load(file)
    except Exception as exc:
        obj = None
        print(exc)
    return obj
```

```python
pipeline_con_data = load_object('../data/pipeline_on_con_records.list')
pipeline_wait_data = load_object('../data/pipeline_on_wait_records.list')
truck_records = load_object('../data/truck_records.list')
package_records = load_object('../data/package_records.list')
```

```python
table_trucks = pd.DataFrame.from_records(truck_records, 
    columns=['truck_id', 'start_wait', 'start_serve', 'end_serve', ])
table_packages = pd.DataFrame.from_records(package_records,
    columns=['package_id', 'start_wait', 'start_serve', 'end_serve', ])
```
