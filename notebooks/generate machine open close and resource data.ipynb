```python
r_a_table_ind = \
    resource_limit_table['resource_name'].str.startswith('a') | resource_limit_table['resource_name'].str.startswith('r')

r_table_ind = \
    resource_limit_table['resource_name'].str.startswith('r')
```

```python
r_table = resource_limit_table[r_table_ind]
```

```python
table = r_table.sample(10)
```

```python
table_sample = table[['resource_id', 'resource_limit']]
```

```python
table_sample['time_stamp'] = '23:00'
```

```python
table_new = table_sample.append(table_sample.assign(
                        resource_limit=1, 
                        time_stamp='21:30',
                                       ))
```

```python
table_new.reset_index(inplace=True, drop=True)
```

```python
write_mysql(table_name='resource_timetable', data=table_new)
```

```python
equipment_timetable = \
    equipment_table[equipment_table.equipment_id.str.startswith('a') | equipment_table.equipment_id.str.startswith('r')]

equipment_timetable = equipment_timetable.sample(10)[['equipment_port',]]

equipment_timetable['time_stamp'] = '23:00'
equipment_timetable['on_off'] = True



equipment_timetable = equipment_timetable.append(equipment_timetable.assign(
                                                        time_stamp='21:30',
                                                        on_off=False,))

equipment_timetable.reset_index(inplace=True, drop=True)
```
