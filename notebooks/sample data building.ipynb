```python
import pandas as pd
import numpy as np
import os

import pickle
```

```python
DATA_DIR = '../inbox'
FILE_NAME = '仿真数据库20170706.xlsx'
DATA_PATH = os.path.join(DATA_DIR, FILE_NAME)

sheet_names = \
    ['i_container_capacity',
    'i_equipment_io',
    'i_equipment_parameter',
    'i_equipment_resource',
    'i_queue_io',
    'i_reload_setting',
    'i_resource_limit', 
    'i_od_parcel_landside',
    'i_unload_setting',
    'i_reload_setting',]
```

```python
tables = {name : pd.read_excel(DATA_PATH, sheetname=name) for name in sheet_names}
```

```python
for name, table in tables.items():
    table.to_csv(f'../data/{name}.csv', index=0)
```

# io_dict
# queue_delay_dict

- i_queue_io
- i_equipement_io

```python
df_io_rules = tables['i_queue_io'].rename(columns={'process_time': 'delay_time', }).merge(
    tables['i_equipment_io'].drop('allocate_rule', 1),
    left_on='queue_in_port',
    right_on='equipment_out_port', how='left')
```

```python
df_io_rules.head(10)
```

```python
import collections

io_dict = collections.defaultdict(list)
queue_delay_dict = dict()

for _, row in df_io_rules.iterrows():
    
    key = row['equipment_in_port']
    val = row['queue_out_port']
    delay = row['delay_time']
    
    io_dict[key].append(val)
    queue_delay_dict[key] = delay
```

# resource_table

- i_equipment_resource
- i_resource_limit

```python
resource_id_dict = dict()

for _, row in tables['i_equipment_resource'].iterrows():
    
    key = row['equipment_in_port']
    val = row['resource_id']
    
    resource_id_dict[key] = val
```

```python
resource_limit_dict = dict()

for _, row in tables['i_resource_limit'].iterrows():
    
    key = row['resource_id']
    val = row['resource_number']
    
    resource_limit_dict[key] = val
```
