```python
import os
import redis
from os.path import join, splitext
import sys
sys.path.append('..')


from src.db import *
```

```python
class RedisConfig:
    HOST = 'localhost'
    PORT = 6379
    DB = 0
    CONN = redis.StrictRedis(host=HOST, port=PORT, db=DB)
```

```python
def reading_tables():
    table_names = os.listdir(SaveConfig.OUT_DIR)
    tables_dict = dict()
    for name in table_names:
        tables_dict[splitext(name)[0]] = pd.read_csv(join(SaveConfig.OUT_DIR, name))
    return tables_dict
```

```python
tables_dict = reading_tables()
```

```python
table_pipeline = tables_dict['pipeline_table']
```

```python
%time table_pipeline_s = table_pipeline.to_msgpack()
```

```python
RedisConfig.CONN.setnx('pipeline_tables', table_pipeline_s)
```

```python
table_pipeline_s_get_ = RedisConfig.CONN.get('pipeline_table')
```
