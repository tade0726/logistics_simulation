```python
c_ports_combos = list(product(['c9_1'], [f'c{i}_{j}' for i, j in product(range(13, 19), range(1, 29))]))
data = pd.DataFrame(c_ports_combos, columns=['equipment_port_last', 'equipment_port_next'])

queue_id = \
    data.equipment_port_last.str.split('_').str.get(0) + '_' + data.equipment_port_next.str.split('_').str.get(0)

data = data.assign(
                equipment_type='c',
                queue_id=queue_id,
                process_time=900.,
                allocate_rule=None,
                parcel_type='Small',
                sorter_zone=None,
                dest_type='A',
                normal_path=0,
                inserted_on=np.datetime64('NaT'),
                modified_on=np.datetime64('NaT'),
)

data = data[table.columns]
```
